**IMPORTANDO BIBLIOTECAS**

```python
from matplotlib import pyplot as plt
import pytesseract
import cv2
```

**CONFIGURANDO O PATH DO TESSERACT (configure de acordo com seu computador)**

```python
pytesseract.pytesseract.tesseract_cmd = 'C:\\Users\\uriel\\AppData\\Local\\Programs\\Tesseract-OCR\\tesseract.exe'
```

**LENDO A IMAGEM**

```python
imagem_bgr = cv2.imread('./images/imagem_borrada.jpg')
imagem_rgb = cv2.cvtColor(imagem_bgr, cv2.COLOR_BGR2RGB)

plt.xticks([])
plt.yticks([])
plt.title('Imagem Original')
plt.imshow(imagem_rgb)
```

**FAZENDO OCR NA IMAGEM ORIGINAL**

```python
texto_imagem_original = pytesseract.image_to_string(imagem_rgb)
print(texto_imagem_original)
```

**REMOVENDO RUÍDOS DA IMAGEM ORIGINAL**

```python
imagem_cinza = cv2.cvtColor(imagem_rgb, cv2.COLOR_RGB2GRAY)
imagem_otimizada = cv2.adaptiveThreshold(imagem_cinza, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 255, 5)

cv2.imwrite('./images/imagem_otimizada.jpg', imagem_otimizada)

plt.xticks([])
plt.yticks([])
plt.title('Imagem Otimizada')
plt.imshow(imagem_otimizada, cmap='gray')
```

**FAZENDO OCR NA IMAGEM OTIMIZADA**

```python
texto_imagem_otimizada = pytesseract.image_to_string(imagem_otimizada)
print(texto_imagem_otimizada)
```
